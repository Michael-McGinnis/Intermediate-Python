## Collections

```python
#Counter, namedtuple, OrderedDict, defaultdict, deque

from collections import Counter
from collections import namedtuple 
from collections import OrderedDict
from collections import defaultdict
from collections import deque 

x = 'qqqqqqwwwwwwwweeeeeeeee'
my_counter = Counter(x)
print(my_counter)

#How many of each element
print(my_counter.values())

#Most common element
print(my_counter.most_common(1))

print(list(my_counter.elements()))

#Create a class with namedtuple called Bucket with values a & b
Bucket = namedtuple('Bucket', 'a, b')

#Add values to a & b
bckt = Bucket(-10, 10)
print(bckt)

#Ordered dict will hold values & key in memory as inputted
ordered_dict = OrderedDict()
ordered_dict['a'] = 1
ordered_dict['b'] = 2
ordered_dict['c'] = 3
ordered_dict['d'] = 4

print(ordered_dict)

ordered_dict['c'] = 3
ordered_dict['d'] = 4
ordered_dict['a'] = 1
ordered_dict['b'] = 2

print(ordered_dict)

d = defaultdict(int)
d['a'] = 1
d['b'] = 2
print(d['a'])

d = deque()

d.append(1)
d.append(2)
d.appendleft(3)
print(d)

d.pop()
d.popleft()
print(d)

d.extend([4, 5, 6])
print(d)

d.rotate(2)
print(d)
```

## Itertools

```python
#Itertools: product, permutations, combinations, accumulate, groupby
#and infinite iterators

from itertools import product
from itertools import permutations
from itertools import combinations
from itertools import accumulate
from itertools import groupby
import operator

x = [1, 2]
y = [3, 4]

prod = product(x, y)
print(prod)
print(list(prod))

a = [1, 2, 3]
perm = permutations(a)
#Can also add a number in permutations for a specified length
#perm = permutations(a, 2)
print(list(perm))

#Combinations

comb = combinations(a, 2)
print(list(comb))

#The function above will not show repetitious results:
# import combinations, combinations_with_replacement
#comb = combinations_with_replacements(a, 2)

#Groupby

def smaller_than_5(x):
    return x < 5

a = [1, 2, 3, 4, 5, 6]
group_obj = groupby(a, key = smaller_than_5)
try:
    for key, value in group_obj:
        print(key, list(value))
except:
    print(False, list(value))
    
#We can also use the lambda function

def smaller_than_5(x):
    return x < 5

a = [1, 2, 3, 4, 5, 6]
group_obj = groupby(a, key = lambda x: x<5)
try:
    for key, value in group_obj:
        print(key, list(value))
except:
    print(False, list(value))

#Accumulate

a = [1, 2, 3, 4]
accum = accumulate(a)
print(a)
print(list(accum))

#Accumulate by default computes the sums
#We can multiply by using the operator

accum = accumulate(a, func = operator.mul)
print(a)
print(list(accum))
```

## Lambda

```python
# Lambda is an anonymous one line function

add = lambda x: x + 10
print(add(5))

#This is the same as:

def add_func(x):
    return x + 10

multiply = lambda x, y: x + y
print(multiply(2, 7))

example = [(1, 2), (12, -1), (6, 8), (-1, -9)]
example_sorted = sorted(example)

print(example)
print(example_sorted)

example = [(1, 2), (12, -1), (6, 8), (-1, -9)]
#Using lambda we can sort using the second element
example_sorted = sorted(example, key = lambda x: x[1])

print(example)
print(example_sorted)

#Sort by each tuples sum
example_sorted = sorted(example, key = lambda x: x[0]+ x[1])
print(example_sorted)

#Map takes a function and applies it to each element
a = [1, 2, 3, 4, 5]

b = map(lambda x: x + 2, a)
print(list(b))

#List comprehension also can do the same
#b = [x + 2 for x in a]

from functools import reduce

product_a = reduce(lambda x, y: x * y, a)
print(product_a)
```

## JSON

```python
# JavaScript Object Notation
# JSON is a standard text-based format for representing structured data 
# based on JavaScript object syntax. It is commonly used for transmitting 
# data in web applications (e.g., sending some data from the server to the client,
# so it can be displayed on a web page, or vice versa)

import json

individual = {"name": "David", "age": 30, "city": "Knoxville",
              "has_children": False, "titles": ["programmer",
                                                "data analyst"]}
individual_JSON = json.dumps(individual)

print(individual_JSON)

individual_JSON = json.dumps(individual, indent = 4, sort_keys = True)
#dumps, s = string

print(individual_JSON)

with open('individual.JSON', 'w') as file:
           json.dump(individual, file, indent = 4)
        
# Convert from JSON back to dictionary
with open('individual.JSON', 'r') as file:
    individual = json.loads(individual_JSON)
    print(individual)
```
